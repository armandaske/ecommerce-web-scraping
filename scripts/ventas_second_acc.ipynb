```python
#Cantidad de productos vendidos
conteo_ventas = ventas_jefe.groupby('Título de la publicación').Unidades.sum()
conteo_ventas = pd.DataFrame(conteo_ventas)
conteo_ventas['Costo unitario sin iva']=pd.Series(costos_dir)
conteo_ventas['Total']=conteo_ventas['Costo unitario sin iva']*conteo_ventas['Unidades']
#extras= pd.read_excel(r'C:\Users\Dell-G3\Documents\Drive\Mercado Libre and stuff\Registros de ventas\ventas_jefe - copia.xlsx', encoding='UTF-8', header=1, index_col='Producto',usecols=['Producto','Costo unitario sin iva', 'Unidades', 'Total'] )
#conteo_ventas = pd.concat([conteo_ventas,extras])
conteo_ventas = conteo_ventas.groupby(conteo_ventas.index).agg({'Unidades': sum,
                                'Costo unitario sin iva': 'first',
                                'Total':sum}) #juntar en un sólo index todos los productos iguales
conteo_ventas.loc['Total',:]=conteo_ventas.sum()
```

```python
df_ventas = pd.read_csv(r"C:\Users\Dell-G3\Documents\Drive\Mercado Libre and stuff\datos_clientes_junio.csv",delimiter=';', header=0,encoding="utf-8")
```

```python
columns_to_drop=['Número de operación de Mercado Pago (operation_id)',
       'Detalle del estado de la operación (status_detail)',
       'Tipo de operación (operation_type)',
       'Documento de la contraparte (buyer_document)']
ventas=df_ventas.drop(columns_to_drop,axis='columns')
ventas.dropna(how='all',inplace=True,axis=1) #Elimina las columnas que tienen TODAS sus entradas como NaN
ventas.head()
```

```python
#Checar en dónde hay NaNs
ventas.isna().any()
```

```python
#Reemplazar los NaNs con 'approved' y quitar todos los status de refund y charge_back
ventas['Estado de la operación (status)'].fillna('approved',inplace=True)
ventas=ventas[ventas['Estado de la operación (status)']=='approved']
```

```python
#Quitar ventas sin teléfono y sin e-mail
ventas=ventas[(ventas['Teléfono de la contraparte (counterpart_phone_number)'].notna())]
#ventas=ventas[~((ventas['Teléfono de la contraparte (counterpart_phone_number)'].isna())&(ventas['E-mail de la contraparte (counterpart_email)'].isna()))]
```

```python
#Dejar sólo los productos que me interesan
ventas['Descripción de la operación (reason)'].unique()
ventas=ventas[ventas['Descripción de la operación (reason)'].isin(['Cubreboca Plástico De Acrílico (caja/10pzas)','Cubreboca Plástico De Acetato Caja Con 10pzas',
       'Cubreboca Mascarilla De Plástico Acrílico Caja C/5pzas',
       'Cubrebocas Plástico De Acrílico Reutilizable Caja C/5 Pzas'])]
```

```python
#juntar cuánto me compró cada cliente
monto_total_por_cliente=ventas.groupby(by='Nombre de la contraparte (counterpart_name)')['Valor del producto (transaction_amount)'].sum()
monto_total_por_cliente.name='monto total de compras'
```

```python
#Quitar la columna de valor de compra individual y poner la de monto total de compra
ventas.drop('Valor del producto (transaction_amount)',axis=1,inplace=True)
ventas=pd.concat([ventas.set_index('Nombre de la contraparte (counterpart_name)'),monto_total_por_cliente],join='inner',axis=1)
```

```python
#Dejar solo los clientes con compras iguales o mayores a $540 (es el monto de 2 paquetes de 5pzas)
ventas=ventas[ventas['monto total de compras']>=540]
```

```python
#Quitar telefonos repetidos
ventas=ventas.drop_duplicates(subset='Teléfono de la contraparte (counterpart_phone_number)')
```

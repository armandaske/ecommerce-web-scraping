```python
#Cantidad de productos vendidos
conteo_ventas = ventas_jefe.groupby('Título de la publicación').Unidades.sum()
conteo_ventas = pd.DataFrame(conteo_ventas)
conteo_ventas['Costo unitario sin iva']=pd.Series(costos_dir)
conteo_ventas['Total']=conteo_ventas['Costo unitario sin iva']*conteo_ventas['Unidades']
#extras= pd.read_excel(r'C:\Users\Dell-G3\Documents\Drive\Mercado Libre and stuff\Registros de ventas\ventas_jefe.xlsx', encoding='UTF-8', header=1, index_col='Producto',usecols=['Producto','Costo unitario sin iva', 'Unidades', 'Total'] )
#conteo_ventas = pd.concat([conteo_ventas,extras])
conteo_ventas = conteo_ventas.groupby(conteo_ventas.index).agg({'Unidades': sum,
                                'Costo unitario sin iva': 'first',
                                'Total':sum}) #juntar en un sólo index todos los productos iguales
conteo_ventas.loc['Total',:]=conteo_ventas.sum()
```

```python
#Cálculo de deuda
deuda_bruta=conteo_ventas.loc['Total','Total']
jefe_debe = 15000+1850+1850+2200+1160+180+245+300+100 #PRESTAMO, TENENCIA, TENENCIA, COSTURERA, BATAS, GORROS, CASTORES, CASTORES, PRESTAMO
debo = deuda_bruta-jefe_debe
print(('Deuda total:%f')%round(deuda_bruta,2))
print(('Deuda Jefe:%f')%round(jefe_debe,2))
print(('DEBO:%f')%round(debo,2))
conteo_ventas
```
